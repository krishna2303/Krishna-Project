```python
import glob
import dlib
import cv2
import random
import numpy as np
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
face_model = dlib.face_recognition_model_v1('dlib_face_recognition_resnet_model_v1.dat')
```

```python
def face_vector(input_path):
    input_image = cv2.imread(input_path)
    faces = detector(input_image, 1)
    if not faces:
        return None
    f = faces[0]
    shape = predictor(input_image, f)
    face_descriptor = face_model.compute_face_descriptor(input_image, shape)
    return face_descriptor
```

```python
f1 = 0 # Chaitanya
f2 = 1 # Unknow
```

```python
print("Retrieving CHAITANYA images ...")
face1=glob.glob("CHAITANYA\*.jpg")
print("Retrieved {} faces !".format(len(face1)))
print("Retrieving unknown images ...")
unknown = glob.glob("Unknown\*.jpg")
print("Retrieved {} faces !".format(len(unknown)))
```

```python
vectors = dlib.vectors()
labels = dlib.array()
```

```python
vectors
```

```python
face1
```

```python
print("Reading Chaitanya images ...")
for i, path in enumerate(face1):
    print("Reading {} of {}\r".format(i, len(face1)))
    face_vectors = face_vector(path)
    if face_vectors is None:
        continue
    vectors.append(dlib.vector(face_vectors))
    labels.append(f1)
```

```python
print("Reading UnKnown  images ...")
for i, path in enumerate(unknown):
    print("Reading {} of {}\r".format(i, len(unknown)))
    face_vectors = face_vector(path)
    if face_vectors is None:
        continue
    vectors.append(dlib.vector(face_vectors))
    labels.append(f2)
```

# Training

```python
type(vectors)
```

```python
vec = np.array(vectors)
```

```python
vec.shape
```

```python
lab = np.array(labels)
```

```python
lab
```

## Test Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(vec,lab,test_size = 0.2, random_state = 0)
```

## Building Model

```python
import keras
```

```python
#Sequential --To init NN model
from keras.models import Sequential
```

```python
#Layers
from keras.layers import Dense
```

#### Init the NN model

```python
model=Sequential()
```

#### Add Input Layer

```python
x_train.shape[1]
```

```python
#input_dim--No of neurons we want in this layer

#kernel_initializer --Kernel --weight, Strategy we want to use to initliaze weights

#activation --Activation function we want to apply

#units --No of ouputs from neurons in current layer or No of neurons in next layer

model.add(Dense(input_dim=x_train.shape[1], kernel_initializer="random_uniform",activation="relu",units=20))
```

#### Add Hidden Layer

```python
model.add(Dense(kernel_initializer="random_uniform",activation="relu",units=15))
```

#### Add Output Layer

```python
model.add(Dense(kernel_initializer="random_uniform",activation="sigmoid",units=1))
```

#### Compile the Model

```python
#In compilation we will mention the strategies to be followed by our NN will Trining 
#loss --Error calculating strategy
#binary_crossentropy  -(ylog(p)+(1-y)log(1-p))

#optimizer --Weights updation strategy or Model optimization strategy
#adam --Mini batch gradient descent

#metrics --Model performance measurment strategy

model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
```

```python
plot_model(model
          )
```

# Train the Model

```python
#epochs--When entire training data is passed that make an epoch
train_history = model.fit(x_train,y_train, epochs=10,batch_size=32,validation_data=(x_test,y_test))
```
