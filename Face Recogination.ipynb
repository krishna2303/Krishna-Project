```python
import cv2
```

```python
import dlib
```

```python
import imutils
```

```python
from imutils import face_utils
```

```python
detector=dlib.get_frontal_face_detector()
```

```python
image=cv2.imread("nm.jfif")
```

```python
#image
#Resolution
rect=detector(image,1)
```

```python
type(rect)
```

```python
rect
```

```python
rect=rect[0]
```

```python
rect
```

```python
face_utils.rect_to_bb(rect)
```

```python
x1=rect.bl_corner().x
y1=rect.tl_corner().y
x2=rect.br_corner().x
y2=rect.br_corner().y
```

```python
x1,y1,x1,y2
```

```python
image=cv2.rectangle(image,(x1,y1),(x2,y2),(255,0,0),2)
```

```python
cv2.imshow("Image",image)
cv2.waitKey()
```

```python
Predictor=dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
#Image
#Rect
shape=Predictor(image,rect)
```

```python
#shape to Numpy
facial_landmarks=face_utils.shape_to_np(shape)
```

```python
len(facial_landmarks)
```

```python
for x,y in facial_landmarks:
    image=cv2.circle(image,(x,y),2,(0,0,255),1)
```

```python
cv2.imshow("Image",image)
cv2.waitKey()
```

```python
video=cv2.VideoCapture(0)
while True:
    data,frame=video.read()
    rect=detector(frame,1)
    for face in rect:
        shape=Predictor(frame,face)
        facial_landmarks=face_utils.shape_to_np(shape)
        for x,y in facial_landmarks:
            cv2.circle(frame,(x,y),2,(0,255,197),2)
    cv2.imshow("Image",frame)
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
video.release()
```
